```python
from anser import *
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
patrick_BL = boundaryLayer( distDomainLims = [ 1e-1 , 1e3 ] )
```

```python
patrick_BL.viscousPropertiesDefine( 15.0e-6 , 986.5e-3 )
```

```python
patrick_BL.boundaryLayerLimitDefine( [ None , 7.99e-3 , 6.31e-3 ] , 27.0 )
```

```python
# Load the image
background_image = mpimg.imread('./imgs/vanDriestDist.png')

# Create a figure
fig, ax = plt.subplots()

# Since imshow doesn't directly support log scale, we convert the x-axis to log scale manually
ax.imshow(background_image, extent=[-1, 3, 0, 25], aspect='auto')

# Step 4: Overlay your plot
x_data = np.log10( patrick_BL.ypluss )
ax.plot( x_data , patrick_BL.Upluss , 'k-', label='Data')  # Example plot


# Add BL points
ax.axvline( x = np.log10( 6.31e-3 * patrick_BL.shearVelocity / patrick_BL.kineticViscosity ) , label = "Momentum Height" )
ax.axvline( x = np.log10( 7.99e-3 * patrick_BL.shearVelocity / patrick_BL.kineticViscosity ) , label = "Displacement Height" )

ax.set_xlim((-1,3))
ax.legend()
ax.set_title("Law of the wall distribution comparison")

```

```python
plt.semilogx( patrick_BL.H_sweep , patrick_BL.error_sweep * 100 , label = "Net Error" )
plt.semilogx( patrick_BL.H_sweep , patrick_BL.disp_error_sweep * 100 , label = "Displacement Error" )
plt.semilogx( patrick_BL.H_sweep , patrick_BL.mom_error_sweep * 100 , label = "Momentum Error" )

plt.grid( which = 'both' , linestyle = '--' , linewidth = 0.5 , color = 'gray' )
plt.grid( which = 'minor' , linestyle = ':' , linewidth = 0.5 , color = 'gray' )

plt.ylim((-100,100))
plt.ylabel( "Error [%]" )
plt.xlabel( "H [m]" )

plt.legend( loc = 'best' )
plt.title("Error of H values")
```
